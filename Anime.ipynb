```python
# Episodes Extraction

def extract_episodes(text):
    check = False
    data=""
    for i in text:
        if i == ")":
            break
        if check == True:
            data+= i
        if i == '(':
            check = True
        
    return data
```

```python
df['Episodes'] = df['Title'].apply(extract_episodes)  
```

```python
df["Episodes"]=(df['Episodes'].str.replace(" eps", "")).astype(int)
```

```python
df.loc[1]
```

```python
# Timestamp Extraction
def extract_timestamp(text):
    count = 0
    data=""
    for i in text:
        if i == ")":
            data= text[count+1:count+20]
            break
        count += 1
    return data
```

```python
df['Total Time'] = df['Title'].apply(extract_timestamp)
```

```python
# Clear Title Name
def clear_name(text):
    return text.split('(')[0].strip()
```

```python
df['Title'] = df['Title'].apply(clear_name)
```

```python
df
```

```python
# Get total months of runtime
from datetime import datetime

def total_runtime(text):
    start_str, end_str = text.split(' - ')
    
    start = datetime.strptime(start_str, "%b %Y")
    end = datetime.strptime(end_str, "%b %Y")
    
    return (end.year - start.year) * 12 + (end.month - start.month) + 1

df['Months'] = df['Total Time'].apply(total_runtime)
```
